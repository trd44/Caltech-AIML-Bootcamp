```python
for col in df_categorical:
    null_vals = df_categorical[col].isna().sum(axis=0)
    if null_vals:
        print(col, " has ", null_vals, " null value(s)")
```

### Fill in the missing values with NA which is what they should have been accoridng to the data description file

```python
for col in df_categorical:
    null_vals = df_categorical[col].isna().sum(axis=0)
    if null_vals:
        df_categorical[col] = df_categorical[col].fillna('NA')
```

```python
for col in df_categorical:
    null_vals = df_categorical[col].isna().sum(axis=0)
    if null_vals:
        print(col, " has ", null_vals, " null value(s)")
```
